# Lab 2.3: Twin RAG Pipelines - Complete Solution

Side-by-side comparison of Ollama and watsonx RAG systems.

```python
import time
import pandas as pd
import matplotlib.pyplot as plt

# Mock comparison results
results = [
    {'question': 'What is RAG?', 'ollama_time': 1.2, 'watsonx_time': 2.1, 'ollama_quality': 8, 'watsonx_quality': 9},
    {'question': 'Explain vectors', 'ollama_time': 1.1, 'watsonx_time': 2.0, 'ollama_quality': 7, 'watsonx_quality': 9},
    {'question': 'What is Granite?', 'ollama_time': 1.3, 'watsonx_time': 2.2, 'ollama_quality': 7, 'watsonx_quality': 10}
]

df = pd.DataFrame(results)

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Speed
ax1.bar(['Ollama', 'watsonx'], [df['ollama_time'].mean(), df['watsonx_time'].mean()])
ax1.set_ylabel('Avg Response Time (s)')
ax1.set_title('Speed Comparison')

# Quality
ax2.bar(['Ollama', 'watsonx'], [df['ollama_quality'].mean(), df['watsonx_quality'].mean()])
ax2.set_ylabel('Quality Score (1-10)')
ax2.set_title('Quality Comparison')

plt.tight_layout()
plt.savefig('comparison.png')
print('âœ… Analysis complete!')
print(df)
```
