## ğŸ¤– Section 3: Creating Your First AI Agent

### What is a CrewAI Agent?

An **Agent** in CrewAI is an autonomous AI entity with:
- A specific **role** (its function in the team)
- A clear **goal** (what it's trying to achieve)
- A **backstory** (context that shapes its behavior)
- An **LLM** (the brain that powers it)

### Agent Design Principles

1. **Single Responsibility**: Each agent should have one clear purpose
2. **Clear Goals**: Goals should be specific and measurable
3. **Rich Backstory**: Provides context for better decision-making
4. **Appropriate Tools**: Agents work best with the right tools

### Creating a Research Agent

````python
from crewai import Agent

# Create a research agent powered by watsonx.ai
research_agent = Agent(
    role="Senior AI Research Specialist",

    goal=(
        "Conduct thorough research on technical topics and produce "
        "well-structured, accurate documentation that serves as a "
        "foundation for content creation."
    ),

    backstory=(
        "You are an experienced AI researcher with a PhD in Computer Science "
        "and 10 years of industry experience at IBM. You excel at breaking down "
        "complex technical concepts into clear, organized notes. Your research "
        "is always thorough, accurate, and well-cited. You have a talent for "
        "identifying the most important information and presenting it in a way "
        "that others can easily understand and build upon."
    ),

    llm=watsonx_llm,  # Powered by watsonx.ai!
    verbose=True,
    allow_delegation=False,
)

print("âœ… Research Agent Created!")
print("="*70)
print(f"Role: {research_agent.role}")
print(f"Goal: {research_agent.goal}")
print("Powered by: IBM watsonx.ai Granite 3 8B")
print("="*70)
````

## ğŸ‘¥ Section 4: Building a Multi-Agent Workflow

### The Power of Agent Collaboration

Multi-agent systems excel at complex tasks by:
- **Dividing work** among specialized agents
- **Building on each other's outputs**

### Our Content Creation Crew

We'll build a three-agent system that works together to create high-quality content:

```
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”

       â–¼
```

### Step 1: Define All Agents

````python
from crewai import Agent, LLM

# Create LLM instances for different agents
# You can use different temperatures for different agents!

# Research agent - Lower temperature for focused, factual responses
researcher_llm = LLM(
    model="watsonx/ibm/granite-3-8b-instruct",
    api_key=os.environ["WATSONX_API_KEY"],
    base_url=os.environ["WATSONX_URL"],
    project_id=os.environ["WATSONX_PROJECT_ID"],
    temperature=0.2,  # Very focused
    max_tokens=1024,
)

# Writer agent - Medium temperature for creativity with structure
writer_llm = LLM(
    model="watsonx/ibm/granite-3-8b-instruct",
    api_key=os.environ["WATSONX_API_KEY"],
    base_url=os.environ["WATSONX_URL"],
    project_id=os.environ["WATSONX_PROJECT_ID"],
    temperature=0.5,  # Balanced
    max_tokens=1536,
)

# Editor agent - Lower temperature for precision
editor_llm = LLM(
    model="watsonx/ibm/granite-3-8b-instruct",
    api_key=os.environ["WATSONX_API_KEY"],
    base_url=os.environ["WATSONX_URL"],
    project_id=os.environ["WATSONX_PROJECT_ID"],
    temperature=0.3,  # Focused but flexible
    max_tokens=2048,
)

# Agent 1: Researcher
researcher = Agent(
    role="Senior AI Research Specialist",
    goal=(
        "Conduct comprehensive research on technical topics and produce "
        "well-organized, accurate documentation."
    ),
    backstory=(
        "You are an experienced AI researcher with deep expertise in enterprise "
        "AI, multi-agent systems, and IBM technologies. You excel at gathering "
        "comprehensive information, identifying key concepts, and organizing them "
        "into clear, structured notes. Your research serves as the foundation for "
        "high-quality content creation."
    ),
    llm=researcher_llm,
    verbose=True,
)

# Agent 2: Technical Writer
writer = Agent(
    role="Senior Technical Writer",
    goal=(
        "Transform research notes into engaging, practical tutorial content "
        "that technical audiences can easily understand and apply."
    ),
    backstory=(
        "You are a skilled technical writer with 8 years of experience creating "
        "developer documentation and tutorials. You have a gift for explaining "
        "complex technical concepts in clear, accessible language. You understand "
        "how to structure content for maximum comprehension and engagement, and you "
        "always include practical examples that readers can use immediately."
    ),
    llm=writer_llm,
    verbose=True,
)

# Agent 3: Editor
editor = Agent(
    role="Content Quality Editor",
    goal=(
        "Review and enhance content for clarity, accuracy, structure, and "
        "pedagogical effectiveness, ensuring it's workshop-ready."
    ),
    backstory=(
        "You are a meticulous editor with a background in technical education. "
        "You have taught hundreds of workshops and know exactly what makes content "
        "effective for learning. You focus on improving clarity, fixing errors, "
        "enhancing structure, and ensuring content is accessible to the target audience. "
        "You're especially vigilant about security best practices and practical usability."
    ),
    llm=editor_llm,
    verbose=True,
)

print("âœ… All agents created with custom LLM configurations!")
print("="*70)
print(f"1. {researcher.role} (temp=0.2)")
print(f"2. {writer.role} (temp=0.5)")
print(f"3. {editor.role} (temp=0.3)")
print("="*70)
print("\nğŸ­ Your multi-agent crew is ready!")
````

### Step 2: Define Tasks and Dependencies

````python
from crewai import Task

# Define the topic
TOPIC = "Building Enterprise AI Solutions with IBM watsonx.ai and Multi-Agent Systems"

print(f"ğŸ“š Workshop Topic: {TOPIC}\n")

# Task 1: Research
research_task = Task(
    description=(
        f"Research the topic: '{TOPIC}'\n\n"
        "Your research should include:\n"
        "1. Overview of IBM watsonx.ai platform and its enterprise capabilities\n"
        "2. Introduction to multi-agent systems and their benefits\n"
        "3. How watsonx.ai powers multi-agent workflows\n"
        "4. 8-10 practical enterprise use cases\n"
        "5. Architecture patterns and best practices\n"
        "6. Security and governance considerations\n\n"
        "Format: Well-structured markdown with clear sections."
    ),
    expected_output=(
        "A comprehensive research document (800-1000 words) in markdown format "
        "with sections for each topic area."
    ),
    agent=researcher,
)

# Task 2: Writing
writing_task = Task(
    description=(
        "Using the research notes, write a tutorial-style article.\n\n"
        "The article should:\n"
        "1. Start with an engaging introduction explaining the value proposition\n"
        "2. Explain the architecture and how components work together\n"
        "3. Provide a step-by-step guide for getting started\n"
        "4. Include practical tips and common pitfalls to avoid\n"
        "5. End with next steps and advanced topics\n\n"
        "Target: 1000-1500 words, easy to follow, professional tone."
    ),
    expected_output=(
        "A complete tutorial article in markdown format with clear headings, "
        "short paragraphs, and practical examples."
    ),
    agent=writer,
    context=[research_task],
)

# Task 3: Editing
editing_task = Task(
    description=(
        "Polish the draft article for a professional technical audience.\n\n"
        "Focus on:\n"
        "1. Improving clarity and flow\n"
        "2. Fixing any technical inaccuracies or unclear explanations\n"
        "3. Enhancing structure and readability\n"
        "4. Making security and governance practices explicit\n"
        "5. Ensuring examples are clear and actionable\n"
        "6. Adding helpful callouts or tips where appropriate\n\n"
        "The final output should be publication-ready."
    ),
    expected_output=(
        "A polished, professional article ready for publication or workshop use. "
        "Error-free, well-structured, and highly educational."
    ),
    agent=editor,
    context=[writing_task],
)

print("âœ… All tasks defined!")
print("="*70)
print("Pipeline:")
print("  Research â†’ Gather information")
print("  Writing  â†’ Create tutorial")
print("  Editing  â†’ Polish and finalize")
print("="*70)
````

### Step 3: Create and Execute the Crew

> â±ï¸ **Note**: This will take 3-5 minutes to complete.

````python
from crewai import Crew, Process

# Assemble the crew
content_crew = Crew(
    agents=[researcher, writer, editor],
    tasks=[research_task, writing_task, editing_task],
    process=Process.sequential,
    memory=False,
    embedder=None,
    verbose=True,
)

print("ğŸš€ Starting content creation workflow...")
print("="*70)
print("All agents are powered by IBM watsonx.ai Granite models!")
print("This will take a few minutes. Please be patient!")
print("="*70)
print()

# Execute the crew
result = content_crew.kickoff()

# Extract final content
final_article = str(result)

print("\n\n" + "="*70)
print("ğŸ‰ CONTENT CREATION COMPLETED!")
print("="*70)
print("\nğŸ“„ Final Article:\n")
print(final_article)
print("\n" + "="*70)
print(f"\nğŸ“Š Article Length: {len(final_article)} characters")
print(f"ğŸ“Š Estimated Words: ~{len(final_article.split())} words")
````

### Save the Generated Article

````python
# Save to file
output_filename = "watsonx_multi_agent_article.md"

with open(output_filename, "w", encoding="utf-8") as f:
    f.write(final_article)

print(f"âœ… Article saved to: {output_filename}")
print(f"ğŸ“ You can download this file from the Colab file browser.")
````

## ğŸ’¼ Section 5: Real-World Use Case - Customer Support System

Let's build a practical multi-agent customer support automation system.

### Use Case: Automated Customer Inquiry Handler

**Problem**: Customer support teams are overwhelmed with repetitive inquiries

**Solution**: A three-agent system that:
1. **Classifies** the inquiry type
2. **Generates** an appropriate response
3. **Reviews** quality before sending

### Implementation

````python
from crewai import Agent, Task, Crew, Process, LLM

# Create specialized LLMs for support system
support_llm = LLM(
    model="watsonx/ibm/granite-3-8b-instruct",
    api_key=os.environ["WATSONX_API_KEY"],
    base_url=os.environ["WATSONX_URL"],
    project_id=os.environ["WATSONX_PROJECT_ID"],
    temperature=0.3,
    max_tokens=1024,
)

# Agent 1: Classifier
classifier = Agent(
    role="Customer Inquiry Classifier",
    goal="Accurately categorize customer inquiries for efficient routing",
    backstory=(
        "You are an expert in customer service operations with 10 years of experience. "
        "You can quickly identify the type and urgency of customer inquiries. You classify "
        "inquiries into: Technical Support, Billing, Product Information, Complaints, or "
        "General Questions. You also assess urgency levels."
    ),
    llm=support_llm,
    verbose=True,
)

# Agent 2: Responder
responder = Agent(
    role="Customer Service Response Specialist",
    goal="Generate professional, helpful, and empathetic customer responses",
    backstory=(
        "You are a seasoned customer service professional known for your clear, friendly "
        "communication. You excel at addressing customer concerns with empathy while "
        "providing actionable solutions. You always maintain a professional yet warm tone, "
        "and you know when to escalate issues to human agents."
    ),
    llm=support_llm,
    verbose=True,
)

# Agent 3: Quality Reviewer
reviewer = Agent(
    role="Response Quality Assurance Specialist",
    goal="Ensure all customer responses meet quality standards before sending",
    backstory=(
        "You are a quality assurance expert in customer communications. You review responses "
        "for professionalism, clarity, completeness, empathy, and accuracy. You catch errors, "
        "suggest improvements, and ensure responses align with company policies. You have a "
        "keen eye for tone and customer satisfaction."
    ),
    llm=support_llm,
    verbose=True,
)

print("âœ… Customer Support System agents created!")
print("="*70)
print("Agents: Classifier â†’ Responder â†’ Reviewer")
print("All powered by watsonx.ai Granite models")
print("="*70)
````

### Test the Support System

Let's process a sample customer inquiry through our multi-agent system.

````python
# Sample customer inquiry
customer_inquiry = """
I've been trying to access my account for the past two days but keep getting
an 'Authentication Failed' error. I've reset my password twice and cleared my
browser cache, but nothing works. This is urgent because I need to submit an
important project by tomorrow. Can someone help me ASAP?
"""

print("ğŸ“§ Processing Customer Inquiry")
print("="*70)
print(f"Inquiry: {customer_inquiry.strip()}\n")

# Task 1: Classify
classify_task = Task(
    description=f"Classify this customer inquiry and assess urgency:\n{customer_inquiry}",
    expected_output=(
        "Classification category, urgency level, and brief reasoning. "
        "Format: Category | Urgency | Reason"
    ),
    agent=classifier,
)

# Task 2: Generate Response
respond_task = Task(
    description=(
        f"Based on the classification, generate a professional response to:\n{customer_inquiry}\n\n"
        "The response should:\n"
        "- Acknowledge the issue and show empathy\n"
        "- Provide clear troubleshooting steps\n"
        "- Set appropriate expectations\n"
        "- Offer escalation if needed\n"
    ),
    expected_output="A complete, professional customer service response email",
    agent=responder,
    context=[classify_task],
)

# Task 3: Review Quality
review_task = Task(
    description=(
        "Review the customer response for quality. Check:\n"
        "1. Professionalism and tone\n"
        "2. Completeness and accuracy\n"
        "3. Empathy and customer focus\n"
        "4. Clarity of instructions\n\n"
        "If approved, output the final response. If improvements needed, suggest changes."
    ),
    expected_output="Final approved response ready to send to customer",
    agent=reviewer,
    context=[respond_task],
)

# Create support crew
support_crew = Crew(
    agents=[classifier, responder, reviewer],
    tasks=[classify_task, respond_task, review_task],
    process=Process.sequential,
    verbose=True,
)

# Process the inquiry
result = support_crew.kickoff()

print("\n" + "="*70)
print("ğŸ“¤ FINAL APPROVED CUSTOMER RESPONSE")
print("="*70)
print(result)
print("="*70)
print("\nâœ… Response ready to send!")
````

### Essential Code Patterns

#### Create LLM Instance
```python
from crewai import LLM

llm = LLM(
    model="watsonx/ibm/granite-3-8b-instruct",
    api_key=os.environ["WATSONX_API_KEY"],
    base_url=os.environ["WATSONX_URL"],
    project_id=os.environ["WATSONX_PROJECT_ID"],
    temperature=0.3,
    max_tokens=1024,
)
```

#### Create Agent
```python
from crewai import Agent

agent = Agent(
    role="Your Role",
    goal="Your Goal",
    backstory="Your Backstory",
    llm=llm,
    verbose=True,
)
```

#### Create Task
```python
from crewai import Task

task = Task(
    description="What to do",
    expected_output="Expected format",
    agent=agent,
    context=[previous_task],  # Optional
)
```

#### Create and Run Crew
```python
from crewai import Crew, Process

crew = Crew(
    agents=[agent1, agent2],
    tasks=[task1, task2],
    process=Process.sequential,
    verbose=True,
)

result = crew.kickoff()
print(result)
```
