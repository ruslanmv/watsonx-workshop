```python
import getpass
from ibm_watsonx_ai import Credentials
from ibm_watsonx_ai.foundation_models import Model
from ibm_watsonx_ai.metanames import GenTextParamsMetaNames as GenParams

model = Model(
    model_id="ibm/granite-3-8b-instruct",
    credentials=Credentials(url="https://us-south.ml.cloud.ibm.com", api_key=getpass.getpass("API Key: ")),
    project_id=getpass.getpass("Project ID: "),
    params={GenParams.MAX_NEW_TOKENS: 250}
)
print("‚úÖ Email responder ready")
```

```python
def auto_respond(email_body: str, category: str = "general") -> dict:
    prompt = f"""You are a professional customer service assistant. Generate a helpful, professional email response.

Customer Email:
{email_body}

Category: {category}

Response:"""
    
    response = model.generate_text(prompt=prompt)
    return {
        "original": email_body,
        "category": category,
        "response": response.strip()
    }
```

```python
# Mock customer emails
incoming_emails = [
    {"body": "Hi, I need to reset my password but the link isn't working. Can you help?", "category": "technical_support"},
    {"body": "What are your business hours and do you have weekend support?", "category": "general_inquiry"},
    {"body": "I'd like to cancel my subscription effective immediately. How do I proceed?", "category": "billing"},
    {"body": "The product I received is damaged. I need a replacement.", "category": "returns"}
]

for i, email in enumerate(incoming_emails, 1):
    print(f"\n{'='*80}\nüì® Email {i} ({email['category']})\n{'='*80}")
    result = auto_respond(email['body'], email['category'])
    print(f"\nüì• Customer:\n{result['original']}")
    print(f"\nüì§ Auto-Response:\n{result['response']}\n")
```

```python
# Try your email
your_email = input("Enter customer email: ")
category = input("Category (general/technical/billing/returns): ") or "general"

result = auto_respond(your_email, category)
print(f"\n{'='*80}\n‚úâÔ∏è GENERATED RESPONSE\n{'='*80}\n\n{result['response']}\n")
```
